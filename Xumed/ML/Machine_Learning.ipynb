```python
# imports 
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## APPLE

```python
# load 
aapl_df = pd.read_csv('/content/drive/MyDrive/class_/week_20/final_aapl.csv')
aapl_df.head()
```

```python
aapl_df.info()
```

```python
aapl_df.isnull().sum()
```

```python
# drop nulls
aapl_df = aapl_df.dropna()
```

```python
aapl_df.isnull().sum()
```

```python
# drop date?
#aapl_df = aapl_df.drop(columns=['Date'], inplace=True)
#aapl_df.head()
```

```python
# convert to datetime
aapl_df['Date'] = pd.to_datetime(aapl_df['Date'])
```

```python
aapl_df.dtypes
```

```python
# APPLE's histogram
col = aapl_df[['tweet_count', 'price_action', 'Volume']]
for i in col.columns:
  plt.hist(col[i])
  plt.title(i)
  plt.show()
```

```python
import seaborn as sns 

print(col.corr())
sns.heatmap(col.corr())
```

```python
# Realtionship between tweet count and price action

plt.scatter(aapl_df.tweet_count, aapl_df.price_action)
plt.xlabel('tweets')
plt.ylabel('price action')
plt.show()
```

```python
X = aapl_df.tweet_count.values.reshape(-1,1)
```

```python
X.shape
```

```python
y = aapl_df.price_action
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
# Realtionship between tweet count and Volume

plt.scatter(aapl_df.tweet_count, aapl_df.Volume)
plt.xlabel('tweets')
plt.ylabel('Volume')
plt.show()
```

```python
y = aapl_df.Volume

model.fit(X, y)

y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```
